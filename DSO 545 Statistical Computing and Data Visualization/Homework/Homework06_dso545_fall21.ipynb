#### 1. Merge both data files (`population.csv` and `codes.csv`) in order to create a new column in the population dataset for the codes of the states.

```python
# data manipulation
import pandas as pd
import numpy as np
# plotly
import plotly.express as px
# dashboards
import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_table
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
from datetime import date
import plotly.graph_objects as go
```

```python
df1 = pd.read_csv('population.csv')
df2 = pd.read_csv('statecodes.csv')
df = df1.merge(df2,how = 'inner',right_on = 'State',left_on = 'state').drop('state',axis = 1)
```

#### 2. Using plotly (or plotly express), create an interactive map that shows the population of the different states in the US for the year 2010.

- Plot title: Population of the USA in 2010
- Color scale: 'Reds'

```python
fig = px.choropleth(df,
                    locations='Code',
                    color='Population',
                    color_continuous_scale='Reds',
                    hover_name='State',
                    locationmode='USA-states',
                    title = 'Population of the USA in 2010',
                    labels={'Population':'Population'},
                    scope='usa')
fig.show()
```
