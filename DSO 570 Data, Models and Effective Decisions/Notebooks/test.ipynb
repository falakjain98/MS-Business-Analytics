```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
import pandas as pd
import sys

rng=default_rng(seed=0)
csvFile='tmp_test.csv'
xlsxFile='tmp_test.xlsx'
df=pd.DataFrame({'X':np.linspace(0,10,10),'Random Data 1':norm(0,1).rvs(10),\
               'X':np.linspace(0,10,10),'Random Data 2':rng.binomial(8,.6,10)})
df.to_csv(csvFile,index=False)
df=pd.read_csv(csvFile)
df.to_excel(xlsxFile,sheet_name='Test',index=False)
df=pd.read_excel(xlsxFile,sheet_name='Test')
df.plot('X',kind='line')
plt.show()

print('Success (Python %d.%d)'%(sys.version_info[0],sys.version_info[1]))
```
